```python
df1 = pd.read_csv("./서울시 기상관련변수1.csv", parse_dates=['일시'])
df=pd.read_csv("./서울시 기상관련변수 20240101 ~ 20241231_기상청.csv", parse_dates=['일시'])
```

```python
start_date = pd.Timestamp("2024-01-01 00:00:00")
end_date = pd.Timestamp("2024-12-31 23:00:00")
```

```python
df1_rest = df1[(df1['일시']<start_date)|(df1['일시']>end_date)]
df_rest = df[(df['일시']>=start_date)&(df['일시']<=end_date)]
```

```python
merged = pd.concat([df1_rest, df_rest]).sort_values('일시').reset_index(drop=True)
```

```python
merged.to_csv("서울시 기상관련변수1.csv",index=False)
```
