```python
df.rename(columns={
    '지점': 'station_id',
    '지점명' : 'station_name',
    '일시' : 'datetime',
    '기온(°C)' : 'temperature',
    '풍향(deg)' : 'wind_direction',
    '풍속(m/s)' : 'wind_speed',
    '강수량(mm)' : 'precipitation',
    '습도(%)' : 'humidity'
}, inplace=True)


# 결과 확인
print(df.columns)
```

```python
import json
from sqlalchemy import create_engine 
from sqlalchemy.exc import SQLAlchemyError
import cryptography

with open('C:/Users/user-pc/Documents/ai-proj2-class4/10.data-collecting/db-config.json', 'r') as f:
    config = json.load(f)

user = config['user']
password = config['password']
host = config['host']
port = config['port']
database = config['database']

# df['datetime'] = pd.to_datetime(df['datetime'])  # datetime 형식으로 변환


```

```python
# SQLAlchemy 엔진 생성
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")
conn = engine.connect()
# 데이터 삽입 (예외처리 포함)
try:
    df.to_sql(name='weather', con=conn, if_exists='append', index=False)
    print("[Success] 데이터가 성공적으로 삽입되었습니다.")
except SQLAlchemyError as e:
    print("[Failure] 데이터 삽입 중 오류 발생:", str(e))
```
