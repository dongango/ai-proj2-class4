```python
def convert_date(val):
    year = int(val)
    month = int(round((val - year) * 100))
    return pd.Timestamp(f"{year}-{month:02d}-01")

df['datetime'] = df['datetime'].apply(convert_date)
```

```python
dust = pd.read_csv('china_aod_daily.csv')
dust.head()
```

```python
dust.rename(columns={"AOD": "aod_avg"}, inplace=True)
dust.rename(columns={"date": "datetime"}, inplace=True)
dust.head()
```

```python
with open('db-config.json', 'r') as f:
    config = json.load(f)

user = config['user']
password = config['password']
host = config['host']
port = config['port']
database = config['database']

dust = dust.copy() 
dust['datetime'] = pd.to_datetime(dust['datetime'])  # datetime 형식으로 변환
```

```python
# SQLAlchemy 엔진 생성
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")
conn = engine.connect()
# 데이터 삽입 (예외처리 포함)
try:
    dust.to_sql(name='china_yellow_dust', con=conn, if_exists='append', index=False)
    print("[Success] 데이터가 성공적으로 삽입되었습니다.")
except SQLAlchemyError as e:
    print("[Failure] 데이터 삽입 중 오류 발생:", str(e))
```

```python
df_max = pd.read_csv('china_aod_max_daily.csv')
df_max.rename(columns={"date": "datetime"}, inplace=True)
df_max.rename(columns={"AOD_max": "aod_max"}, inplace=True)
df_max
```

```python
df_db = pd.read_sql("SELECT id, datetime FROM china_yellow_dust", con=engine)
df_db['datetime'] = pd.to_datetime(df_db['datetime'])
```

```python
df_db['datetime'] = pd.to_datetime(df_db['datetime'])
df_max['datetime'] = pd.to_datetime(df_max['datetime'])
```

```python
update_df = pd.merge(df_db, df_max, how='left', left_on='datetime', right_on='datetime')
update_df = update_df[['id', 'aod_max']].dropna()
update_df
```

```python
from sqlalchemy import text

with engine.begin() as conn:
    for _, row in update_df.iterrows():
        conn.execute(
            text("UPDATE china_yellow_dust SET aod_max = :aod_max WHERE id = :id"),
            {"aod_max": row['aod_max'], "id": int(row['id'])}
        )
```
