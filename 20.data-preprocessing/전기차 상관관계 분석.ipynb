```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. CSV  
df = pd.read_csv(r'C:\Users\USER\Documents\ai-proj2-class4\20.data-preprocessing\_.csv', encoding='utf-8')

# 2.     
cols = ['no2', 'co', 'so2', 'o3', 'pm10', 'pm25', 'diesel', 'electric', 'gasoline']
corr_df = df[cols]

# 3.   (Spearman:  ,   )
corr_matrix = corr_df.corr(method='spearman')

# 4.  
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='Blues', fmt=".2f", square=True)
plt.title('Spearman  ', fontsize=15)
plt.tight_layout()
plt.show()

import matplotlib as mpl

#  
mpl.rcParams['font.family'] = 'Malgun Gothic'
#    
mpl.rcParams['axes.unicode_minus'] = False
```
