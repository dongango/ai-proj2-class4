```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. CSV 파일 불러오기
df = pd.read_csv(r'C:\Users\USER\Documents\ai-proj2-class4\20.data-preprocessing\전기차용_데이터셋.csv', encoding='utf-8')

# 2. 상관관계 분석 대상 열만 선택
cols = ['no2', 'co', 'so2', 'o3', 'pm10', 'pm25', 'diesel', 'electric', 'gasoline']
corr_df = df[cols]

# 3. 상관계수 계산 (Spearman: 순위 기반, 비선형 관계에도 강건)
corr_matrix = corr_df.corr(method='spearman')

# 4. 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='Blues', fmt=".2f", square=True)
plt.title('Spearman 상관관계 행렬', fontsize=15)
plt.tight_layout()
plt.show()

import matplotlib as mpl

# 폰트 설정
mpl.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 기호 깨짐 방지
mpl.rcParams['axes.unicode_minus'] = False
```
